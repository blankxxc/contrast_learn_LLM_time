```python
data=pd.read_csv('./AAPL_reduce_text3.csv')
data2=np.log(data.iloc[:,1:-1]).diff()
data2
```

```python
data2.columns
```

```python
# å°†å·®åˆ†åŽçš„æ•°æ®è¿›è¡Œåˆ†ç®±ï¼Œæ¯ä¸ªç®±å­åˆ’åˆ†ä¸º0.01çš„æ•°å­—
bin_width = 0.01
bins = np.arange(data2["Open"].min()-bin_width, data2["Open"].max() + bin_width, bin_width)
list_logdiff = pd.cut(data2["Open"].values, bins=bins)  

# ç»˜åˆ¶ç›´æ–¹å›¾
plt.figure(figsize=(20, 12))
list_logdiff.value_counts().sort_index().plot(kind='bar', color='skyblue', edgecolor='black')
plt.title('Histogram of Log Difference')
plt.xlabel('Log Difference Bins')
plt.ylabel('Frequency')
plt.show()
```
